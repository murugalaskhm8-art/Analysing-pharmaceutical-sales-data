```python
import pandas as pd

df = pd.read_csv(r"Analysing-pharmaceutical-sales-data\Data\salesdaily.csv")

df.head()
```

```python
total_sales = df.iloc[:,1:9].sum()

print(total_sales.sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt

total_sales.sort_values().plot(
    kind='barh',
    figsize=(10,5)
)

plt.title("Total Sales Quantity by Drug Category")
plt.xlabel("Total Sales")
plt.show()
```

```python
top_drug = total_sales.idxmax()

print("Highest Selling Drug Category:", top_drug)
```

```python
highest_drug = total_sales.idxmax()
highest_sales = total_sales.max()

print("Highest Selling Drug:", highest_drug)
print("Sales:", highest_sales)
```

```python
total_sales.sort_values(ascending=False).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Highest Selling Drug Category")
plt.xlabel("Drug Category")
plt.ylabel("Sales")
plt.show()
```

```python
drug_cols = ['M01AB', 'M01AE', 'N02BA', 'N02BE', 'N05B', 'N05C', 'R03', 'R06']
```

```python
jan2015 = df[(df['Year'] == 2015) & (df['Month'] == 1)]

top3 = jan2015[drug_cols].sum().sort_values(ascending=False).head(3)

print(top3)
```

```python
top3.plot(kind='bar')

plt.title("Top 3 Drugs - January 2015")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    top3,
    labels=top3.index,
    autopct='%1.1f%%'
)

plt.title("Top 3 Drugs - July 2016")
plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    top3,
    labels=top3.index,
    autopct='%1.1f%%'
)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Top 3 Drugs - September 2017")
plt.show()
```

```python
sales2017 = df[df['Year']==2017][drug_cols].sum()

sales2017.sort_values(ascending=False).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Drug Sales in 2017")
plt.ylabel("Sales")
plt.show()
```

```python
avg_sales = df[drug_cols].mean()

avg_sales.sort_values(ascending=False).plot(
    kind='line',
    marker='o',
    figsize=(10,5)
)

plt.title("Average Daily Sales")
plt.ylabel("Average Sales")
plt.grid(True)
plt.show()
```

```python
r03_monthly = df.groupby('Month')['R03'].mean()

r03_monthly.plot(
    kind='line',
    marker='o',
    figsize=(10,5)
)

plt.title("Monthly Average Sales of R03")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.grid(True)
plt.show()
```
